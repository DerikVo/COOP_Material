````python
%%sql
with initial_suspects_Cte AS (
SELECT --explaining to COOPers how to convert things into a CTE
  id
FROM
  Person
WHERE
    (address_street_name = 'Northwestern Dr'
    AND address_number = (SELECT MAX(address_number) FROM Person WHERE address_street_name = 'Northwestern Dr')) --explain to coopers how to get the max house number for this street
  OR
  ( Name like '%Annabel%'
    AND address_street_name = 'Franklin Ave')
)


SELECT
    i.person_id
    ,p.license_id
    ,i.transcript
FROM
  interview as i
  JOIN
  Person as p on i.person_id = p.id
WHERE
  i.person_id in (SELECT * FROM initial_suspects_Cte) --Explaining to COOPers how to use a sub-query
````

````python
%%sql
  SELECT -- we find potential suspects at the gym, all thats left is to find which member matches the plate number
    *
  FROM
    get_fit_now_member
  WHERE
    id like ('48Z%')
    AND membership_status = 'gold'
````

````python
%%sql
SELECT
  p.id as person_id
  ,p.license_id as Drivers_license
  ,g.name as Name
  ,d.plate_number
  ,g.id as gym_id
FROM
  Person as p
  JOIN
  get_fit_now_member as g on p.id = g.person_id
  JOIN
  drivers_license as d on d.id = p.license_id
WHERE
  g.name in (  SELECT name
                FROM
                  get_fit_now_member
                WHERE
                  g.id like ('48Z%')
                  AND g.membership_status = 'gold'
                  AND d.plate_number like '%H42W%') -- we get the out answer by examing the plate number
````

To make it easier to read we can convert the subqery into a CTE

````python
%%sql
WITH gym_suspect AS ( --Creates A CTE to find the information about the Gym suspect
  SELECT
    name
  FROM
    get_fit_now_member
  WHERE
    g.id LIKE '48Z%' --This information is found by interviewing the suspects caught on camera
    AND g.membership_status = 'gold'
    AND d.plate_number LIKE '%H42W%'
)

SELECT
  p.id as person_id
  ,p.license_id as Drivers_license
  ,g.name as Name
  ,d.plate_number
  ,g.id as gym_id
FROM
  Person as p -- We join various tabls together to get all the information
  JOIN
    get_fit_now_member as g on p.id = g.person_id
  JOIN
    drivers_license as d on d.id = p.license_id
WHERE
  --uses a subquery
  g.name IN (SELECT name FROM gym_suspect); -- We use a subquery to pull from CTE
````

````python
%%sql
WITH concert_cte AS (
    SELECT
      p.id
    FROM
      person as p
      JOIN
        facebook_event_checkin AS f  ON f.person_id = p.id
    WHERE
      event_name = 'SQL Symphony Concert'
      AND date LIKE '201712%'
    GROUP BY
      p.id
    HAVING
      COUNT(*) = 3
)

SELECT
  *
FROM
  concert_cte AS cc
  JOIN
  person as p ON cc.id = p.id
  JOIN drivers_license as dl ON dl.id = p.license_id
WHERE
  gender = 'female'
  AND height BETWEEN 65 AND 67
  AND hair_color = 'red'
  AND car_make = 'Tesla'
  AND car_model = 'Model S';
````
