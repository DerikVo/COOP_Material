````python
%%sql
SELECT
  *
FROM
  crime_scene_report
WHERE
  date= 20180115
  AND
  city = 'SQL City'
  AND
  Type = 'murder'
````

**Derik notes:**
Next we look at the crimereport to find that we have two witnesses. One lives on the last house on Northwestern Dr, and the other person is nammed Annabel and lives somewhere on Franklin Ave.

Here I use a CTE to get the ID of our people in question. In the CTE I nested where clauses to get each individual. With in my nested id for Northwestern Dr, I use a subquery to get the highest address number on northwestern dr. This makes an assumption that house numbers go in acending order.

Next I use the CTE and query from our interview table to filter own our suspects. Feel free to run the CTE as its own indivdual query to see what the main query is using to find our interview notes.

````python
%%sql
with initial_suspects_Cte AS (
SELECT --explaining to COOPers how to convert things into a CTE
  id
FROM
  Person
WHERE
    (address_street_name = 'Northwestern Dr'
    AND address_number = (SELECT MAX(address_number) FROM Person WHERE address_street_name = 'Northwestern Dr')) --explain to coopers how to get the max house number for this street
  OR
  ( Name like '%Annabel%'
    AND address_street_name = 'Franklin Ave')
)


SELECT
    i.person_id
    ,p.license_id
    ,i.transcript
FROM
  interview as i
  JOIN
    Person as p on i.person_id = p.id
WHERE
  i.person_id in (SELECT * FROM initial_suspects_Cte) --Explaining to COOPers how to use a sub-query
````

**Derik notes:**

With our witness testimony we get information from a potential suspect. We use this information to find our potential suspect.


Here I use a basic query to find potential suspects based solely on gym membership information, but I want to be sure our suspect matches all our evidence.

````python
%%sql
  SELECT -- we find potential suspects at the gym, all thats left is to find which member matches the plate number
    *
  FROM
    get_fit_now_member
  WHERE
    id like ('48Z%')
    AND membership_status = 'gold'
````

**Derik notes:**

To do this you have to join on the tables that contain the appropriate information. You can do this by using a subquery using the above query and adding on any additional information. In this case we use the plate number of their car.

````python
%%sql
SELECT
  p.id as person_id
  ,p.license_id as Drivers_license
  ,g.name as Name
  ,d.plate_number
  ,g.id as gym_id
FROM
  Person as p
  JOIN
    get_fit_now_member as g on p.id = g.person_id
  JOIN
    drivers_license as d on d.id = p.license_id
WHERE
  g.name in (  SELECT name
                FROM
                  get_fit_now_member
                WHERE
                  g.id like ('48Z%')
                  AND g.membership_status = 'gold'
                  AND d.plate_number like '%H42W%') -- we get our answer by incorporating the plate number
````

**Derik notes:**

To make it easier to read we can convert the subqery into a CTE

````python
%%sql
WITH gym_suspect AS ( --Creates A CTE to find the information about the Gym suspect
  SELECT
    name
  FROM
    get_fit_now_member
  WHERE
    id LIKE '48Z%' --This information is found by interviewing the suspects caught on camera
    AND membership_status = 'gold'

)

SELECT
  p.id as person_id
  ,p.license_id as Drivers_license
  ,g.name as Name
  ,d.plate_number
  ,g.id as gym_id
FROM
  Person as p -- We join various tabls together to get all the information
  JOIN
    get_fit_now_member as g on p.id = g.person_id
  JOIN
    drivers_license as d on d.id = p.license_id
WHERE
  --uses a subquery
  g.name IN (SELECT name FROM gym_suspect) -- We use a subquery to pull from CTE
  AND d.plate_number LIKE '%H42W%' --incorporate their plate number.
````

**Derik notes:**

Here we do the same thing, but this time I will jump stright into creating a CTE to find suspects that match the description given by our suspect.

After that the main query, will pull the other pieces of information such as carl model, height, and hair color. You can do all of this in the CTE and do a select *, but I did it this way so its easier for people to visualize whats happening.

````python
%%sql
WITH concert_cte AS (
    SELECT
      p.id, count(*) as concert_attendance
    FROM
      person as p
      JOIN
        facebook_event_checkin AS f  ON f.person_id = p.id
    WHERE
      event_name = 'SQL Symphony Concert'
      AND date LIKE '201712%'
    GROUP BY
      p.id
    HAVING
      concert_attendance = 3
)

SELECT
  p.id
  ,p.Name
  ,dl.hair_color
  ,dl.height
  ,dl.car_model
  ,cc.concert_attendance
FROM
  concert_cte AS cc
  JOIN
    person as p ON cc.id = p.id
  JOIN
    drivers_license as dl ON dl.id = p.license_id
WHERE
  gender = 'female'
  AND height BETWEEN 65 AND 67
  AND hair_color = 'red'
  AND car_make = 'Tesla'
  AND car_model = 'Model S';
````
